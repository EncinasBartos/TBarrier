# Overview

We illustrate the numerical implementation and the use of active LCS diagnostics on 2-D homogeneous, isotropic turbulence. The notebook is structured as follows:

1. Import precomputed trajectories stored in the folder 'data/Turbulence'.
<br />
2. Define computational parameters (such as the number of cores) and data.
 <br />
3. Define spatio-temporal domain.
 <br />
4. Interpolate velocity from (discrete) gridded data.
 <br />
6. active Finite Time Lyapunov Exponent (aFTLE):

   <br />
   
    1. Instantaneous linear momentum barriers

    * For incompressible, uniform-density Navier–Stokes flows, the instantaneous barrier equation for linear momentum takes the
    specific forms:
    
    \begin{equation}
    \mathbf{x'} = \nu \rho \Delta \mathbf{u}
    \label{eq: linearmoment_1}
    \end{equation}
    
    For 2D incompressible, uniform-density Navier–Stokes flows, the Eulerian barrier equation \ref{eq: linearmoment_1} for linear momentum is an autonomous Hamiltonian system of the form:
    
    \begin{equation}
    \mathbf{x'} = \nu \rho \nabla \mathbf{\omega}(\mathbf{x}, t)
    \label{eq: linearmoment_2}
    \end{equation}
    
    Whenever the vorticity $ \mathbf{\omega}(\mathbf{x}, t) $ is precisely known, it is easier to work with eq. \ref{eq: linearmoment_2} than eq. \ref{eq: linearmoment_1} as it requires one less derivative to compute.
    
    In order to evalutae eq. $ \ref{eq: linearmoment_2} $, we first need to compute $\mathbf{\omega}(\mathbf{x}, t)  $.  
    
    <br />
   
    1. Instantaneous vorticity transport barriers
    
    * For incompressible, uniform-density Navier–Stokes flows, the instantaneous barrier equation for vorticity transport takes the
    specific forms:
    
    \begin{equation}
    \mathbf{x'} = \nu \rho \Delta \mathbf{\omega}
    \label{eq: vorticity_barrier_1}
    \end{equation}
    
    2D incompressible, uniform-density Navier–Stokes flows, the Eulerian barrier equation \ref{eq: vorticity_barrier_1} vorticity transport barriers is an autonomous Hamiltonian system of the form:
    
    \begin{equation}
    \mathbf{x'} = \nu J  \nabla \dfrac{D}{Dt}\mathbf{\omega}(\mathbf{x}, t)
    \label{eq: vorticity_barrier_2}
    \end{equation}
    
    Again, whenever the vorticity $ \mathbf{\omega}(\mathbf{x}, t) $ is precisely known, it is easier to work with eq. \ref{eq: linearmoment_2} than eq. \ref{eq: linearmoment_1} as it requires one less spatial derivative to compute.

# Import Data

```python
%%time
# import mat73 as scipy does not work when importing large mat-files
import mat73

# import scipy
import scipy.io

# import numpy
import numpy as np

# load vorticity data
turb_w = parent_directory + r'/data/Turbulence/turb_w'

omega = np.zeros((1024, 1024, 251))
# iterate over all files in folder 'turb_w' and load data
i = 0
for filename in sorted(os.listdir(turb_w)):
    if filename.endswith(".mat"):
        file_w = os.path.join(turb_w, filename)
        mat_turb_w = scipy.io.loadmat(file_w)
        omega[:,:,i] = mat_turb_w['w']
        i+=1

# load x/y grid
X = scipy.io.loadmat(parent_directory + r'/data/Turbulence/grid.mat')['x']
Y = X.copy().transpose()

#Import trajectory data from file in data-folder
mat_file = mat73.loadmat('../../../../data/Turbulence/Particles1100.mat')

xt = mat_file['xt'] # shape (51, 1099, 1099, 5)
yt = mat_file['yt'] # shape (51, 1099, 1099, 5)

# Number of cores for parallel computing
Ncores = 18
```

# Spatio-temporal domain

```python
# specify time at which we want to compute instantaneous momentum barriers (this should an integer in [0, 251])
t = 0

# specify domain over which to compute linear momentum barriers
x_domain = np.linspace(4.2, 5, 300)
y_domain = np.linspace(2.2, 3, 300)

X_domain, Y_domain = np.meshgrid(x_domain, y_domain)

# Define ratio of auxiliary grid spacing vs original grid_spacing
aux_grid_ratio = .2 # [1/5, 1/100]
aux_grid = [np.around(aux_grid_ratio*(X_domain[0, 1]-X_domain[0, 0]), 5), np.around(aux_grid_ratio*(Y_domain[1, 0]-Y_domain[0, 0]), 5)]
```

# Instantaneous linear momentum barriers

We visualize linear momentum barriers by applying the FTLE-field on the trajectories of system \ref{eq: linearmoment_2}. As soon as the FTLE-field is applied to an active vector field, such as linear momentum, we refer to it as aFTLE (active FTLE). We will subsequently compare the results with the classic version of the instantaneous FTLE-field which is given by the maximum eigenvalue of the rate of strain tensor.

## Interpolate instantaneous momentum barrier field

We start by computing the interpolant for the instantaneous momentum barrier field. The interpolant is given by the right-hand-side of equation \ref{eq: linearmoment_2}.

## Compute active FTLE

We now compute the active FTLE for the instantaneous linear momentum barrier equations. First, we evaluate the active Cauchy-Green strain tensor from the gradient of the active flow map over the meshgrid. The active flow map is given by the solution of the autonomous ODE from equation \ref{eq: linearmoment_2}. Finally, we apply the classic FTLE to the active Cauchy-Green strain tensor.

```python
%%time
# Import interpolation function for instantaneous active momentum barrier equation
from ipynb.fs.defs.InstantaneousActiveMomentum import InstantaneousActiveMomentum

# Interpolate velocity data using cubic spatial interpolation
Interpolant = InstantaneousActiveMomentum(X, Y, omega[:,:,t], aux_grid)
```

```python
%%time
# Import gradient of flow map
from ipynb.fs.defs.ActiveGradientFlowMap import ActiveGradientFlowMap

# Import package which computes eigenvalues/eigenvectors
from ipynb.fs.defs.eigen import eigen

# Import package for progress bar
from tqdm.notebook import tqdm

# Import package for parallel computing
from joblib import Parallel, delayed

# Import package for computing CauchyGreen (CG) strain tensor
from ipynb.fs.defs.CauchyGreen import CauchyGreen

# Import package for computing Finite-Time Lyapunov Exponent (FTLE)
from ipynb.fs.defs.FTLE import _FTLE

# define dummy time interval
ds = 0.001
tN = 0.15
s_dummy = np.arange(0, tN+ds, ds)

# parallel computation of active FTLE over meshgrid.
def parallel_FTLE(i):
    
    Interpolant_x = Interpolant[0]
    Interpolant_y = Interpolant[1]
    
    FTLE_parallel = np.zeros((X_domain.shape[1], 2))*np.nan
    
    for j in range(X_domain.shape[1]):
        
        # set initial condition
        x = np.array([X_domain[i, j], Y_domain[i, j]])%(2*np.pi)
        
        Interpolant_x = Interpolant[0]
        Interpolant_y = Interpolant[1]
    
        # compute active gradient of flow map from finite differencing using auxiliary grid
        activegradFmap = ActiveGradientFlowMap(s_dummy, x, Interpolant, aux_grid)
        
        # activegradFmap has shape (2, 2, len(s_dummy)) --> we need gradient of flow map from [0, 0.15] and from [0, 0.05]
        
        # activegradFmap from [0, 0.05]
        activegradFmap_t0_t05 = activegradFmap[:,:,50]
        
        # activegradFmap from [0, 0.15]
        activegradFmap_t0_t15 = activegradFmap[:,:,-1]
        
        # compute active CG from [0, 0.05]
        C05 = CauchyGreen(activegradFmap_t0_t05)
            
        # compute FTLE from CG
        FTLE_parallel[j, 0] = _FTLE(C05, s_dummy[-1])
        
        # compute active CG from [0, 0.15]
        C15 = CauchyGreen(activegradFmap_t0_t15)
            
        # compute FTLE from CG
        FTLE_parallel[j, 1] = _FTLE(C15, s_dummy[-1])

    return FTLE_parallel

FTLE = np.array(Parallel(n_jobs=Ncores, verbose = 0, backend = "threading", require = "sharedmem")(delayed(parallel_FTLE)(i) for i in tqdm(range(X_domain.shape[0]))))
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (9, 6))
ax = plt.axes()
cax = ax.contourf(X_domain, Y_domain, FTLE[:,:,1], levels = 600, cmap = "winter")
fig.colorbar(cax)
plt.show()
```

```python
fig = plt.figure(figsize = (9, 6))
ax = plt.axes()
cax = ax.contourf(X_domain, Y_domain, FTLE[:,:,0], levels = 600, cmap = "winter")
fig.colorbar(cax)
plt.show()
```
