```python
import numpy as np

def eigen(A, incompressible):
                   
    a = A[0,0]
    b = A[0,1]
    c = A[1,0]
    d = A[1,1]
            
    lamda_max = (a+d)/2+np.sqrt((a+d)**2/4-(a*d-b*c))
        
    if incompressible:
            
        if lamda_max < 1:
                
            lamda_max = np.nan
        
        lamda_min = 1/lamda_max
            
    else:
            
        lamda_min = (a+d)/2-np.sqrt((a+d)**2/4-(a*d-b*c))
        
    if np.abs((a+d)**2/4-(a*d-b*c)) <= 10**(-14):
                       
        if b == 0 and c == 0:
            
            v_max = np.array([1, 0])
            v_min = np.array([0, 1])
                
        elif b != 0 and c == 0:
                
            v_max = np.array([1, 0])
            v_min = np.array([0, 1])
                
        elif b == 0 and c != 0:
                
            v_max = np.array([0, 1])
            v_min = np.array([1, 0])
            
        else:
                
            v_max = np.array([-b, a-lamda_max])
            v_max = v_max/np.sqrt(v_max[0]**2+v_max[1]**2)
            
            v_min = np.array([a-lamda_max, b])
            v_min = v_min/np.sqrt(v_min[0]**2+v_min[1]**2)
            
    else:
        
        if a-lamda_max != 0 and b != 0:
            
            v_max = np.array([-b, a-lamda_max])
            v_max = v_max/np.sqrt(v_max[0]**2+v_max[1]**2)
            
            v_min = np.array([a-lamda_max, b])
            v_min = v_min/np.sqrt(v_min[0]**2+v_min[1]**2)
        
        else:
            
            v_max = np.array([d-lamda_max, -c])
            v_max = v_max/np.sqrt(v_max[0]**2+v_max[1]**2)
            
            v_min = np.array([c, d-lamda_min])
            v_min = v_min/np.sqrt(v_min[0]**2+v_min[1]**2)
                        
    return lamda_min, lamda_max, v_min, v_max
```
